```python
import requests
import json
import pathlib
import os
```

```python
images_dir = os.path.join(os.getcwd(), "images")
os.makedirs(images_dir, exist_ok=True)
```

```python
with open('launches.json', 'r') as f:
    # Load the JSON data
    data = json.load(f)
```

```python
data
```

```python
res=[i['image'] for i in data['results']]
res
```

```python
for i in res:
    try:
        # print(i)
        res=requests.get(i)
        tmp_name= i.split('/')[-1]
        file_name=f'images/{tmp_name}'
        with open(file_name, "wb") as f:
            f.write(res.content)
        print(file_name)
    except Exception as e:
        # Handling the exception
        print("An error occurred:", e)
        
```
